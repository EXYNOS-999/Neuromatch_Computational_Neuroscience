# Matplotlib figures

Matplotlib has [two distinct interfaces](https://matplotlib.org/3.2.1/tutorials/introductory/usage.html#the-object-oriented-interface-and-the-pyplot-interface), which is a common source of confusion.

**Prefer using the object oriented interface, especially for complex figures. Don't mix interfaces.**

```python
#@title Imports and setup
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The `pyplot` interface is similar to MATLAB and uses a series of imperative commands that interact with a hidden state to draw on a "current figure".

```python
x = np.linspace(-1, 1, 100)
plt.figure(figsize=(4, 4))
plt.plot(x, x ** 2, label="f(x)")
plt.xlabel("x")
plt.ylabel("x$^2$")
plt.title("A quadratic")
plt.legend()
```

The "object oriented" interface defines figure and axes objects and then uses methods on those objects to build the plot.

```python
f, ax = plt.subplots(figsize=(4, 4))
ax.plot(x, x ** 2, label="f(x)")
ax.set(
  xlabel="x",
  ylabel="x$^2$",
  title="A quadratic",
)
ax.legend()
```

The object oriented interface is a bit more verbose for simple plots, but it scales much better to more complex figures. It also avoids the abiguity of issuing commands whose result depends on hidden state, which is a major source of confusion for new (and experienced) users.

```python
x, y = np.random.multivariate_normal([0, 0], [(4, .5), (.5, 1)], 1000).T

f = plt.figure(figsize=(9, 3))
gridspec = f.add_gridspec(1, 3)
ax1 = f.add_subplot(gridspec[:2])
ax2 = f.add_subplot(gridspec[2])

ax1.plot(x)
ax1.plot(y)
ax2.scatter(x, y, color=".2", marker=".")

ax1.set(xlabel="Time (s)", ylabel="Signal")
ax2.set(xlabel="x", ylabel="y")

f.tight_layout()
```
